```python
import csv
import math
import numpy as np

wins = []
with open('strength.csv') as f:
    reader = csv.DictReader(f)
    for row in reader:
        wins.append(float(row['win_rate']))

wins = np.array(wins)
wins.shape
```

```python
import matplotlib.pyplot as plt

plt.plot(wins)
plt.ylim(0, 1)
plt.show()
```

This plot shows the relative strength of the model's policy network vs an MCTS benchmark.
